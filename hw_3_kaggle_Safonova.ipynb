```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from pandas.api.types import is_numeric_dtype
from sklearn.compose import ColumnTransformer
```

```python
X = data.drop(['Churn'], axis=1)
y = data.Churn
```

```python
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(drop='first')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)
    ]
)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=36)

pipeline = make_pipeline(preprocessor, LogisticRegressionCV(scoring='roc_auc', refit=True))
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)

roc_auc = roc_auc_score(y_test, y_pred)

print(roc_auc)
```

```python
from catboost import CatBoostClassifier, Pool, cv, metrics
```

```python
cat_boost_model = CatBoostClassifier(iterations=1000,eval_metric='AUC', verbose=100)
cat_boost_model.fit(X_train, y_train, cat_features=cat_cols, eval_set=(X_test, y_test))
cat_boost_model.best_score_
```

```python
y_pred_proba = cat_boost_model.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC-AUC: {roc_auc}")
```
